```python
#下载数据
import requests
import urllib.request
url="https://www.csie.ntu.edu.tw/~cjlin/libsvmtools/datasets/regression/cadata"
#r=requests.get(url)   #方式一：request获得数据内容，传回二进制数据，需要用Python3的BytesIO(r.content)来获取
urllib.request.urlretrieve(url, "cadata")   #方式二：下载数据集


#加载数据：使用sklearn库的load_svmlight_file函数读取数据
from sklearn.datasets import load_svmlight_file
import numpy as np
from io import BytesIO

#X,y=load_svmlight_file(f=BytesIO(r.content,n_features=8) #对应方式一
X,y=load_svmlight_file("cadata",n_features=8)  #对应方式二

#print type(X)
X=X.data.reshape(X.shape)   #很重要！！！X数据格式需转换成nd.array形式
#print("y变换前：",y.shape)
y=np.reshape(y,(y.shape[0],1))  #很重要！！！（354，）需转换成（354,1）,否则后面出错
#print("y变换后：",y.shape)
#print type(X)

#数据处理：使用train_test_split函数切分数据集
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)   #切割数据集

#数据归一化
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
X_train = min_max_scaler.fit_transform(X_train)
y_train = min_max_scaler.fit_transform(y_train)
X_test = min_max_scaler.fit_transform(X_test)
y_test = min_max_scaler.fit_transform(y_test)
#print(X_train.shape)
#print(y_train.shape)
#print(X_test.shape)
#print(y_test.shape)

#超参初始化
alpha=0.01   #alpha:学习率
iterations=200    #iterations:迭代次数

#线性回归与随机梯度下降算法
m, n = np.shape(X_train)
W = np.ones([n,1])     #初始化W全为1或的n*1向量
#print(W.shape)

losses=[]
import random
for ecoph in range(iterations):
    #print(W.reshape(1,n))
    i = random.randint(0, m-1)  #随机选择一个样本进行梯度下降
    #print(W.reshape(1,n))
    #print(X_train[i].shape,y_train[i].shape)
    grad=np.dot(X_train[i].reshape(1,n).T,(np.dot(X_train[i].reshape(1,n),W))-y_train[i].reshape(y_train[0].shape[0],1))
    #print(hypo.shape) 
    G=-alpha*grad  #G:一个样本的梯度
    W+=G    #梯度下降 
    yp=np.dot(X_test,W)    #yp:预测值
    loss=np.average(((yp-y_test)**2)/2)  #损失值
    #print('ecoph:',ecoph,',loss:',loss)
    losses.append(loss)
```

```python
#画出图形
%matplotlib inline
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))
plt.plot(losses,color='red')
plt.title("Linear Regression and SGD")
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.show()
```
